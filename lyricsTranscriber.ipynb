```python
# Install Whisper if not already installed
!pip install -q git+https://github.com/openai/whisper.git

import whisper
from google.colab import files
import torch

# Load the Whisper model (use 'base', 'small', 'medium', 'large' based on your needs)
model = whisper.load_model("large", device="cuda" if torch.cuda.is_available() else "cpu")

# Step 1: Upload the audio file
# The user can click the "Choose Files" button to select an audio file from their local machine.
uploaded = files.upload()

# Step 2: Load the uploaded audio file
# The code assumes the uploaded file will have a filename that can be accessed through the 'uploaded' dictionary.
audio_file = list(uploaded.keys())[0]

# Step 3: Transcribe the audio using Whisper's transcribe method
result = model.transcribe(audio_file)

# Step 4: Print detected language
print(f"Detected language: {result['language']}")

# Step 5: Print the recognized text
print("Transcription: ", result['text'])

# Note: To insert a new audio file, simply re-run the code and upload the new file when prompted.
```
